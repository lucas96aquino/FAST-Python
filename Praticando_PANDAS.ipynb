```python
import pandas as pd
```

```python
df_vacinados = pd.read_csv('vacinados_CERTO CERTO.csv')
display(df_vacinados)
```

```python
df_vacinados.head()
```

```python
df_vacinados.columns
```

```python
# No DataFrame df_dados_vacinacao, selecione todas as entradas de pessoas com
# idade maior que 60 anos (idade > 60) do gênero feminino.

acima_60_feminino = df_vacinados[(df_vacinados['idade'] >= 60) & (df_vacinados['sexo'] == 'FEMININO')]
print(acima_60_feminino[['idade', 'sexo']])
```

```python
# Ordene o DataFrame df_dados_vacinacao com base na coluna "data_vacinacao" em
# ordem decrescente. Exiba as 10 primeiras entradas.

print(df_vacinados.sort_values('data_vacinacao', ascending=False).head(10))
```

```python
# No DataFrame df_dados_vacinacao filtre as pessoas com faixa etária entre
# 25 a 29 anos da cor parda e preta que moram em Recife que
# tomaram a vacina PFIZER.

filtro_3 = df_vacinados[(df_vacinados['faixa_etaria'] == '25 a 29 anos') & (df_vacinados['raca_cor'] == 'PARDA') | (df_vacinados['raca_cor'] == 'PRETA') & (df_vacinados['municipio'] == 'RECIFE') & (df_vacinados['vacina_fabricante'].str.contains('3 -'))]
display(filtro_3)
```

```python
# Exporte o DataFrame df_dados_vacinacao para um novo arquivo CSV chamado
# "dados_vacinacao_filtrados.csv", incluindo apenas as colunas:
# "faixa_etaria", "idade", "sexo", "raca_cor", "municipio" e "data_vacinacao".

filtro_3.to_csv('filtro_3')
```
