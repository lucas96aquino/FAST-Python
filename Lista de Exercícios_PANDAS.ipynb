```python
# EXERCICIO 1
# No DataFrame df_dados_vacinacao, selecione todas as entradas de pessoas com
# idade maior que 60 anos (idade > 60) do gênero feminino.

import pandas as pd

df_vacinados = pd.read_csv('vacinados_CERTO CERTO.csv')

print(df_vacinados.loc[(df_vacinados['sexo'] == 'FEMININO') & (df_vacinados['idade'] > 60)])
```

```python
# EXERCÍCIO 2
# Ordene o DataFrame df_dados_vacinacao com base na coluna "data_vacinacao"
# em ordem decrescente. Exiba as 10 primeiras entradas.

import pandas as pd

df_vacinados = pd.read_csv('vacinados_CERTO CERTO.csv')

df_filtro = df_vacinados.sort_values('data_vacinacao', ascending=False)
print(df_filtro.head(10))
```

```python
# EXERCÍCIO 3

# No DataFrame df_dados_vacinacao filtre as pessoas com faixa etária entre 25 a
# 29 anos da cor parda e preta que moram em Recife que  tomaram a vacina PFIZER.

import pandas as pd

df_vacinados = pd.read_csv('vacinados_CERTO CERTO.csv')

df_filtro = df_vacinados.loc[(df_vacinados['faixa_etaria'] == '25 a 29 anos') &  ((df_vacinados['raca_cor'] == 'PARDA') | (df_vacinados['raca_cor'] == 'PRETA')) & (df_vacinados['vacina_fabricante'].str.startswith('3 -'))]
df_resultado = df_filtro.sort_values('idade', ascending=True)
print(df_resultado)
```

```python
# Exporte o DataFrame df_dados_vacinacao para um novo arquivo CSV chamado
# "dados_vacinacao_filtrados.csv", incluindo apenas as colunas: "faixa_etaria",
#  "idade", "sexo", "raca_cor", "municipio" e "data_vacinacao".

import pandas as pd

df_vacinados = pd.read_csv('vacinados_CERTO CERTO.csv')
df_vacinados_filtrado = df_vacinados[['faixa_etaria', 'idade', 'sexo', 'raca_cor', 'municipio', 'data_vacinacao']]
print(df_vacinados_filtrado)

df_vacinados_filtrado.to_csv('df_vacinados_filtro.csv', encoding = 'utf-8')
```
